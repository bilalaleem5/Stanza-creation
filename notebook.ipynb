```python



import random
# Random module is imported for generating random verses.

# Required for generating random verses.
# This comment reinforces the purpose of importing the random module.

# Import nltk
# The Natural Language Toolkit library is imported for text processing.

import nltk

# Import the ngrams
# Importing the ngrams module from nltk.util for handling n-grams.

from nltk.util import ngrams

# Load the poets' texts
# Reading the poetry files of Shakespeare and Angelou.

shakespeare_poetry = open('william shakespeare.txt', encoding='utf-8').read()

Robert_poetry = open('Robert Frost.txt', encoding='utf-8').read()

# Tokenize the texts

# Tokenizing the texts to break them into words.

shakespeare_tokens = nltk.word_tokenize(shakespeare_poetry)

Robert_tokens = nltk.word_tokenize(Robert_poetry)

# Create N-gram models (unigram, bigram, and trigram)

# Creating unigram, bigram, and trigram models for both poets.


def create_ngram_model(tokens, n):

    # Function to create n-gram models.


    ngram_model = {}

    for i in range(len(tokens) - n + 1):

        ngram = tuple(tokens[i:i+n])

        next_word = tokens[i + n] if i + n < len(tokens) else None

        if ngram in ngram_model:

            ngram_model[ngram].append(next_word)

        else:

            ngram_model[ngram] = [next_word]

    return ngram_model

shakespeare_unigram_model = create_ngram_model(shakespeare_tokens, 1)

shakespeare_bigram_model = create_ngram_model(shakespeare_tokens, 2)

shakespeare_trigram_model = create_ngram_model(shakespeare_tokens, 3)



Robert_unigram_model = create_ngram_model(Robert_tokens, 1)

Robert_bigram_model = create_ngram_model(Robert_tokens, 2)

Robert_trigram_model = create_ngram_model(Robert_tokens, 3)

# Generate verses using the N-gram models with a word limit

# Function to generate verses based on the selected n-gram model.

def generate_verse(ngram_model, n, max_length=7):

    verse = []

    current_ngram = random.choice(list(ngram_model.keys()))



    for _ in range(max_length):

        if current_ngram not in ngram_model:

            break



        next_word = random.choice(ngram_model[current_ngram])



        if next_word is None:

            break



        if next_word != ' ':

            verse.append(next_word)



        current_ngram = current_ngram[1:] + (next_word,)



    return ' '.join(verse)



# Generate stanzas with verses grouped by n-gram

# Generating stanzas with verses for both poets.



stanza_count = 4


for _ in range(stanza_count):

    print("\nStanza - Shakespeare:")

    for n in range(1, 4):

        print(f"{n}-gram:")
        for _ in range(4):

            verse = generate_verse(shakespeare_unigram_model, n)

            verse_with_spaces = ''.join([word for word in verse if word is not None])

            print(verse_with_spaces)



    print("\nRobert frost:")
    for n in range(1, 4):

        print(f"{n}-gram:")

        for _ in range(4):

            verse = generate_verse(Robert_unigram_model, n)

            verse_with_spaces = ''.join([word for word in verse if word is not None])
            print(verse_with_spaces)


```

```python
import random

# Function to generate a verse
def generate_verse(model, n):
    verse = []

    # Generate a random number of words in the range [7...10]
    num_words = random.randint(7, 10)

    # Select the first word randomly from the starting words in the vocabulary
    starting_words = [gram[0] for gram in model if gram[0].isalpha()]

    if not starting_words:
        return "Error: No starting words found in the model."

    current_word = random.choice(starting_words)
    verse.append(current_word)

    # Choose subsequent words until the end of the verse
    for _ in range(num_words - 1):
        next_words = [gram[1] for gram in model if gram[0] == current_word]

        if not next_words:
            break

        current_word = random.choice(next_words)
        verse.append(current_word)

    return ' '.join(verse)


# Function to generate a stanza
def generate_stanza(poet_name, unigram_model, bigram_model, trigram_model):
    print(f"{poet_name} Style Poem:")

    for _ in range(3):
        # Generate three verses with an empty line between them
        for _ in range(4):
            verse = generate_verse(trigram_model, 3)
            print(verse)
        print()

# Ask the user to choose a poet's style
user_choice = input("Choose a poet's style (William Shakespeare or Robert Frost): ")

if user_choice.lower() == 'william shakespeare':
    generate_stanza("William Shakespeare", shakespeare_unigrams, shakespeare_bigrams, shakespeare_trigrams)
elif user_choice.lower() == 'robert frost':
    generate_stanza("Robert Frost", robert_unigrams, robert_bigrams, robert_trigrams)
else:
    print("Invalid choice. Please select either 'William Shakespeare' or 'Robert Frost'.")
```
